```python
def my_position_list(x):
    x = x.replace(' ','')
    my_position_list = x.split(',')
    defense_list = ['CB','LWB','LB','RB','RWB']
    found = 0
    for i in my_position_list:
        if i in defense_list:
            found = 1
            break
        else:
            found = 0
    return found
```

```python
df['player_positions'].apply(my_position_list)
```

```python
df['defense_found'] = df['player_positions'].apply(my_position_list)
```
